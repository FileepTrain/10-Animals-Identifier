```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
data_dir = "../dataset/raw-img"  # Path to the dataset folder
img_size = (224, 224)  # Resize images to this size
batch_size = 32  # Number of images to process at once
```

```python
def load_data(data_dir, img_size):
    images = []
    labels = []
    classes = os.listdir(data_dir)
    
    for label, category in enumerate(classes):
        category_path = os.path.join(data_dir, category)
        for img_file in os.listdir(category_path):
            img_path = os.path.join(category_path, img_file)
            try:
                img = load_img(img_path, target_size=img_size)  # Load and resize image
                img_array = img_to_array(img) / 255.0  # Normalize pixel values
                images.append(img_array)
                labels.append(label)
            except Exception as e:
                print(f"Error loading image {img_path}: {e}")
    
    return np.array(images), np.array(labels), classes
    
images, labels, class_names = load_data(data_dir, img_size)
print(f"Loaded {len(images)} images across {len(class_names)} categories.")
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    images, labels, test_size=0.2, random_state=42
)
y_train = to_categorical(y_train, num_classes=len(class_names))
y_val = to_categorical(y_val, num_classes=len(class_names))
print(f"Training set: {X_train.shape}, Validation set: {X_val.shape}")
```
